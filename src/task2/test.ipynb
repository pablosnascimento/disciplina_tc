```python
import numpy as np
import math

def AND(a,b): 
    return a*b

def OR(a,b): 
    return 1 if a+b else 0

def XOR(a,b): 
    return 1 if ((a == 1 and b == 0) or (a == 0 and b == 1)) else 0

def NOT(a): 
    return 1-a

def NAND(a,b):
    return 1-AND(a,b)

def MAJ(X0,X1,X2):
    firstpair  = AND(X0,X1)
    secondpair = AND(X1,X2)
    thirdpair  = AND(X0,X2)
    temp       = OR(secondpair,thirdpair)
    return OR(firstpair,temp)

#print(MAJ(0,1,1))

def CMP(X0, X1, X2, X3):
    notc = NOT(X2)
    and1 = AND(X0,notc)
    and2 = AND(X0,X2)
    nota = NOT(X0)
    and3 = AND(nota,notc)
    or1 = OR(and2,and3)
    notd = NOT(X3)
    and4 = AND(X1,notd)
    and5 = AND(or1,and4)
    return OR(and1,and5)

#print("CMP: ", CMP(1,1,1,0))

def GT6(a2, a1, a0, b2, b1, b0):
    nota2 = NOT(a2)
    nota1 = NOT(a1)
    #nota0 = NOT(a0)
    notb2 = NOT(b2)
    notb1 = NOT(b1)
    notb0 = NOT(b0)
    and1 = AND(a2,notb2)

    and2 = AND(nota2,notb2)
    and3 = AND(a2,b2)
    and4 = AND(a1,notb1)
    or1 = OR(and2,and3)
    and5 = AND(or1,and4)
    
    and6 = AND(nota1,notb1)
    and7 = AND(a1,b1)
    and8 = AND(a0,notb0)
    or2 = OR(and6,and7)
    and9 = AND(or2,and8)
    and10 = AND(or1,and9)
    
    or3 = OR(and1,and5)
    or4 = OR(or3,and10)
    
    return or4

def GT6_NAND(a2, a1, a0, b2, b1, b0):
    nota2 = NAND(a2,a2)
    nota1 = NAND(a1,a1)
    #nota0 = NAND(a0,a0) nao precisa
    notb2 = NAND(b2,b2)
    notb1 = NAND(b1,b1)
    notb0 = NAND(b0,b0)
    
    temp1 = NAND(a2,notb2)
    nand1 = NAND(temp1,temp1)

    temp2 = NAND(nota2,notb2)
    nand2 = NAND(temp2,temp2)
    temp3 = NAND(a2,b2)
    nand3 = NAND(temp3,temp3)
    temp4 = NAND(a1,notb1)
    nand4 = NAND(temp4,temp4)
    tempor1 = NAND(nand2,nand2)
    tempor2 = NAND(nand3,nand3)
    nandor1 = NAND(tempor1,tempor2)
    temp5 = NAND(nandor1,nand4)
    nand5 = NAND(temp5,temp5)
    
    temp6 = NAND(nota1,notb1)
    nand6 = NAND(temp6,temp6)
    temp7 = NAND(a1,b1)
    nand7 = NAND(temp7,temp7)
    temp8 = NAND(a0,notb0)
    nand8 = NAND(temp8,temp8)
    tempor3 = NAND(nand6,nand6)
    tempor4 = NAND(nand7,nand7)
    tempor5 = NAND(tempor3,tempor4)
    temp9 = NAND(tempor5,nand8)
    nand9 = NAND(temp9,temp9)
    temp10 = NAND(nandor1,nand9)
    nand10 = NAND(temp10,temp10)
    
    tempor6 = NAND(nand1,nand1)
    tempor7 = NAND(nand5,nand5)
    nandor3 = NAND(tempor6,tempor7)
    tempor8 = NAND(nandor3,nandor3)
    tempor9 = NAND(nand10,nand10)
    nandor4 = NAND(tempor8,tempor9)
    
    return nandor4

print("GT6:", GT6(0,0,0,0,0,0))

def tobin(x):
    return (bin(x)[2:]).zfill(6)

def toint(x):
    return int(x,2)

def ValidaGT6():

    print("------ GT6 Real")
    lines = int(math.pow(2,6))
    m = []
    for i in range(lines):
        bin = tobin(i)
        gt6 = GT6(int(bin[0]),int(bin[1]), int(bin[2]), int(bin[3]), int(bin[4]), int(bin[5]))
        gt6NAND = GT6_NAND(int(bin[0]),int(bin[1]), int(bin[2]), int(bin[3]), int(bin[4]), int(bin[5]))
        a = bin[0:3]
        b = bin[3:6]
        res = 0 
        if a > b:
            res = 1
        m.append(bin+" "+ str(gt6) + " " + str(gt6NAND) + " " + str(res))

    for i in range(lines):
        print(m[i])    
```

```python
def SOMA(x0, x1, x2, x3, x4, x5):
    X = [x0,x1,x2,x3,x4,x5]
    size = len(X)
    hsize = int(size/2)

    Carry = [0]*size
    Y = [0]*size
    
    for i in range(hsize):
        tempi = XOR(X[(hsize-1)-i], X[(size-1)-i])
        Y[i] = XOR(tempi, Carry[i])
        ai = AND(X[(hsize-1)-i],X[(size-1)-i])
        bi = AND(tempi, Carry[i])
        Carry[i+1] = OR(ai, bi)

    Y[hsize] = Carry[hsize]
    return Y
```
